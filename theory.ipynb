# Постановка задачи оптимизации Single shooting
Пусть динамика системы задаётся дифференциальным уравнением

$$
\begin{equation}
    \dot{x} = f(t, x, \theta),\; x(t_0) = c_0,
\end{equation}
$$

где $t \in [t_0, t_f]$ - время; $x \in \mathbb{R}^n$ - вектор состояния системы; 
$\theta \in \mathbb{R}^p$ - вектор параметров, которые мы подразумеваем постоянными на промежутке времени $t \in [t_0, t_f]$; 
$c_0 \in \mathbb{R}^n$ - начальное состояние системы; 
$f : \mathbb{R}^n \times \mathbb{R}^p \to \mathbb{R}^n$ - некоторая функция, непрерывно дифференцируема по $x$ и $\theta$.

Положим *неизвестными* и *независимыми* вектор параметров $\theta \in \mathbb{R}^p$ и начальное состояние $c_0 \in \mathbb{R}^n$ системы (1).

На временном интервале $t \in [t_0, t_f]$ имеется конечный набор из $N$ измерений 

$$
y_i = h(x(t_i), \theta) + \varepsilon_i, \quad i \in \overline{1, N},
$$

где $h: \mathbb{R}^n \times \mathbb{R}^p \to \mathbb{R}^m$ - некоторая функция измерений, которая также непрерывно дифференцируема по $x$ и $\theta$;
$y_i \in \mathbb{R}^m$ - $i$-ое измерение;
$\varepsilon_i$ - случайная ошибка $i$-го измерения, $\varepsilon_i \sim \mathcal{N}(0, \sigma_i)$.

Необходимо найти такие параметры $\hat{\theta}$ и начальное условие $\hat{c}_0$, которые минимизируют некоторую целевую функцию 
$ S(\theta, c_0): \ \mathbb{R}^p \times \mathbb{R}^n \to \mathbb{R}$.

Введем запись $x(t | \theta, c_0)$, которую следует понимать как решение системы (1) в момент времени $t$ с начальным условием $x(t_0) = c_0$ и при значении параметров $\theta$.

В качестве целевой функции $S(\theta, c_0)$ рассмотрим квадратичную функцию
$$
S(\theta, c_0) = \frac{1}{2}\sum_{i=1}^N \left\| h\bigl(x(t_i | \theta, c_0), \theta\bigr) - y_i \right\|^2.
$$

Рассмотрим задачу:

$$
\min_{\theta, c_0} S(\theta, c_0)
$$
 
$$
\begin{aligned}
\text{s.t.}\;\;\;\;\;\;\;\;  &\dot{x} = f(t, x, \theta), \\
&x(t_0) = c_0,
\end{aligned}
$$

**где:**
- $x \in \mathbb{R}^n$ - вектор состония системы;
- $f : \mathbb{R}^n \times \mathbb{R}^p \to \mathbb{R}^n$ - функция, определяющая структуру модели;
- $h : \mathbb{R}^n \times \mathbb{R}^p \to \mathbb{R}^m$ - функция наблюдения;
- $y_i \in \mathbb{R}^m$ - наблюдаемые значения в момент $t_i$;
- $\theta  \in \mathbb{R}^p$ - вектор параметров модели;
- $c_0 \in \mathbb{R}^n$ - начальное состояние системы;
- $N$ - количество наблюдений.


## Вариация целевой функции

Рассмотрим вариацию целевой функции $ S(\theta, c_0) $ относительно параметров $\theta$ и начального условия $c_0$
$$
S(\theta + \delta\theta, c_0 + \delta c_0) = 
\frac{1}{2}\sum_{i=1}^N \left\| h\bigl(x(t_i | \theta+\delta\theta, c_0 + \delta c_0), \theta+\delta\theta\bigr) - y_i \right\|^2.
$$

Последовательно рассмотрим вариации. Начнём с вариации решения системы (1):

$$
    x(t | \theta + \delta \theta, c_0 + \delta c_0) \approx 
    x(t | \theta, c_0) + \frac{\partial x(t | \theta, c_0)}{\partial \theta} \delta\theta + \frac{\partial x(t | \theta, c_0)}{\partial c_0} \delta c_0 = 
    x(t | \theta, c_0) + J_\theta(t | \theta, c_0) \delta\theta + J_{c_0}(t | \theta, c_0) \delta c_0,
$$

где мы обозначили
- $\cfrac{\partial x(t | \theta, c_0)}{\partial \theta} = J_\theta(t | \theta, c_0) \in \mathbb{R}^{n \times p}$ - значение якобиана чувствительности системы (1) по параметрам $\theta$ в момент времени $t$, 
- $\cfrac{\partial x(t | \theta, c_0)}{\partial c_0} = J_{c_0}(t | \theta, c_0) \in \mathbb{R}^{n \times n}$ - значение якобиана чувствительности системы (1) по начальному условию $c_0$ в момент времени $t$.

Значения указанных якобианов в моменты времени $t_i \in [t_0, t_f]$ получаются из решений пары матричных *вариационных* дифференциальных уравнений:

$$
\begin{aligned}
    \frac{d}{dt} J_\theta(t | \theta, c_0) &= \frac{d}{dt} \frac{\partial x(t | \theta, c_0)}{\partial \theta} =
    \frac{\partial}{\partial \theta} \frac{d}{dt} x(t | \theta, c_0) = \frac{\partial}{\partial \theta} f(t, x, \theta) =\\
    &= \frac{\partial f(t, x, \theta)}{\partial x} \frac{\partial x(t | \theta, c_0)}{\partial \theta} + \frac{\partial f(t, x, \theta)}{\partial \theta} = \\
    &= f^\prime_x(t, x, \theta) J_\theta(t | \theta, c_0) + f^\prime_\theta(t, x, \theta), \quad J_\theta(t_0 | \theta, c_0) = \frac{\partial x(t_0 | \theta, c_0)}{\partial \theta} = \frac{\partial{c_0}}{\partial{\theta}} = 0, \\[10pt]

    \frac{d}{dt} J_{c_0}(t | \theta, c_0) &= \frac{d}{dt} \frac{\partial x(t | \theta, c_0)}{\partial c_0} =
    \frac{\partial}{\partial c_0} \frac{d}{dt} x(t | \theta, c_0) = \frac{\partial}{\partial c_0} f(t, x, \theta) =\\
    &= \frac{\partial f(t, x, \theta)}{\partial x} \frac{\partial x(t | \theta, c_0)}{\partial c_0} = \\
    &= f^\prime_x(t, x, \theta) J_{c_0}(t | \theta, c_0), \quad J_{c_0}(t_0 | \theta, c_0) = \frac{\partial x(t_0 | \theta, c_0)}{\partial c_0} = \frac{\partial{c_0}}{\partial{c_0}} = I,
\end{aligned}
$$

здесь, при выводе вариационных уравнений, мы воспользовались предположением, что функция $f(t, x, \theta)$ непревыно дифференцируема по состоянию $x$ и параметрам $\theta$.
Интегрирование вариационных уравнений следует выполнять совместно с интегрированием системы (1), таким образом, в будущем мы увидим, что на каждой итерации алгоритма следует решать *расширенную* систему из $n(p + n + 1)$ дифференциальных уравнений.

Вернёмся к вариации целевой функции, теперь рассмотрим вариацию функции измерений, подставив в нее полученные результаты,

$$
\begin{aligned}
    h \bigl(x(t | \theta+\delta\theta, c_0 + \delta c_0), \theta+\delta\theta \bigr) &\approx
    h \bigl(x(t | \theta, c_0) + J_\theta(t | \theta, c_0) \delta\theta + J_{c_0}(t | \theta, c_0) \delta c_0, \theta+\delta\theta \bigr) \approx \\
    &\approx h \bigl(x(t | \theta, c_0), \theta \bigl) + \frac{\partial h}{\partial x} \bigl( J_\theta(t | \theta, c_0) \delta\theta + J_{c_0}(t | \theta, c_0) \delta c_0 \bigr) + \frac{\partial h}{\partial \theta} \delta\theta = \\
    &= h \bigl(x(t | \theta, c_0), \theta \bigl) + \bigr( h^\prime_x J_\theta(t | \theta, c_0) + h^\prime_{\theta} \bigl) \delta\theta + h^\prime_x J_{c_0}(t | \theta, c_0) \delta c_0.
\end{aligned}
$$

Тогда вариацию целевой функции можно записать как
$$
S(\theta + \delta\theta, c_0 + \delta c_0) \approx 
\frac{1}{2}\sum_{i=1}^N \left\| \bigr( h^\prime_x J_\theta(t_i | \theta, c_0) + h^\prime_{\theta} \bigl) \delta\theta + h^\prime_x J_{c_0}(t_i | \theta, c_0) \delta c_0 - \left(y_i - h (x(t_i | \theta, c_0), \theta) \right) \right\|^2.
$$

Из этой записи становится ясно, что наша исходная задача минимизации может быть сведена к некоторой последовательности квадратичных задач относительно вариаций параметров $\delta\theta$ и начальных условий $\delta c_0$.

## Метод Гаусса-Ньютона
На этом этапе наших рассуждений уже становится ясным, что решение исходной задачи оценки параметров $\theta$ и начальных условий $c_0$ в итоге сводится к итерационной процедуре, где на каждой итерации требуется решить квадратичную задачу относительно вариаций $\delta \theta$ и $\delta c_0$. В этом и заключается метод Гаусса-Ньютона.
Итак, пусть на $k$-ой итерации, $k = 0, 1, 2,\ldots$, у нас имеются значения $\theta^k$ и $c_0^k$, рассмотрим задачу

$$
\min_{\delta\theta, \delta c_0} \frac{1}{2}\sum_{i=1}^N \left\| \bigr( h^\prime_x J_\theta(t_i | \theta^k, c_0^k) + h^\prime_{\theta} \bigl) \delta\theta + h^\prime_x J_{c_0}(t_i | \theta^k, c_0^k) \delta c_0 - \left(y_i - h (x(t_i | \theta^k, c_0^k), \theta^k) \right) \right\|^2,
$$

которая является задачей МНК

$$
\min_{\delta p} \frac{1}{2} \left\| {\bf A}(p^k) \delta p - {\bf r}(p^k) \right\|^2,
$$

где мы обозначили $p^k = [\theta^k \quad c_0^k]^\top \in \mathbb{R}^{p+n}$ - обобщенный вектор параметров; $\delta p = [\delta \theta \quad \delta c_0]^\top \in \mathbb{R}^{p+n}$ - обобщенный вектор переменных;

$$
{\bf A}(p^k) =
\begin{bmatrix}
h^\prime_x J_\theta(t_1 | p^k) + h^\prime_{\theta} & h^\prime_x J_{c_0}(t_1 | p^k) \\
\vdots & \vdots \\
h^\prime_x J_\theta(t_N | p^k) + h^\prime_{\theta} & h^\prime_x J_{c_0}(t_N | p^k)
\end{bmatrix} \in \mathbb{R}^{m N \times (p+n)}; \quad
{\bf r}(p^k) = 
\begin{bmatrix}
y_1 - h (x(t_1 | p^k), p^k) \\
\vdots \\
y_N - h (x(t_N | p^k), p^k)
\end{bmatrix} \in \mathbb{R}^{m N}.
$$

Пусть $\delta p^*$ - решение очередной задачи МНК на $k$-ой итерации, тогда обобщенный вектор параметров на следующей итерации

$$
p^{k+1} = p^{k} + \delta p^*.
$$

Таким образом, итеративно решается исходная задача об оценке параметров $\theta$ динамической системы и начальных условий $c_0$.

**В матричной форме:**
Введем обозначения

$$
\mathbf{p} = 
\begin{bmatrix}
\theta \\
c_1 \\
c_2 \\
\vdots \\
c_T \\
\end{bmatrix} \in \mathbb{R}^{K + m \times T}, \quad

\mathbf{r} = 
\begin{bmatrix}
y_1 - h\big(x(t_1, \theta, c_1)\big) \\
y_2 - h\big(x(t_2, \theta, c_1)\big) \\
\vdots \\ 
y_{N_2 - 1} - h\big(x(t_{N_1 - 1}, \theta, c_1)\big) \\
y_{N_2} - h\big(x(t_{N_2}, \theta, c_2)\big) \\
y_{N_2+1} - h\big(x(t_{N_2+1}, \theta, c_2)\big) \\
\vdots \\ 
y_{N_{T}} - h\big(x(t_{N_{T}}), \theta, c_T)\big) \\
y_{N_{T}+1} - h\big(x(t_{N_{T}+1}), \theta, c_T)\big) \\
\vdots \\ 
y_N - h\big(x(t_N, \theta, c_T)\big)
\end{bmatrix}\in \mathbb{R}^{n \times N}, \quad

$$

$$
\Phi^j_{\theta}(t_i) = \frac{\partial h(x(t_i, \theta, c_j))}{\partial x} J^j_{\theta}(t_i),  \;\;\;\; \Phi_{c_j}(t_i) = \frac{\partial h(x(t_i, \theta, c_j))}{\partial x} J_{c_j}(t_i)
$$

Матрицы $J^j_{\theta}(t_i)$ и $J_{c_j}(t_i)$ являются решением задачи Коши на интервале $[t_{N_j}, t_{i}]$  $t_i \in [t_{N_j}, t_{N_{j+1}}]$


 $$
 \frac{\partial }{\partial t}J^j_{\theta}(t) = \frac{\partial }{\partial \theta}f(\theta, x, t) + \frac{\partial f(t,x, \theta)}{\partial x}J^j_{\theta}(t)
 $$

$$
 \frac{\partial }{\partial t}J_{c_j}(t) = \frac{\partial f(t,x, \theta)}{\partial x}J_{c_j}(t)
 $$


$$
 \frac{\partial x(t, \theta)}{\partial t} = f(t, x, \theta)
$$

$$
J_{c_j}(t_0) =  \mathbf{I}
$$

$$
J^j_{\theta}(t_0) =  \mathbf{0}
$$

$$
x(t_0, \theta) = c_j
$$

Аналогично результату полученному для задачи Single Shooting, можем переписать задачу в матричной форме как квадратичную задачу оптимизации

$$
\min_{\delta p^k} S(\delta p^k) = \| H(p^k)\delta p^k - \mathbf{r^k}\|_{\sigma}
$$
 Где 

 
$$
H(p) =
\begin{bmatrix}
\Phi^1_{\theta}(t_1) & \Phi_{c_1}(t_1) & 0 & 0& \dots& 0\\
\Phi^1_{\theta}(t_2) & \Phi_{c_1}(t_2) & 0 & 0 & \dots &0\\
\vdots  & \vdots  & 0& 0 & \dots &0\\
\Phi^1_{\theta}(t_{N_1 - 1}) & \Phi_{c_1}(t_{N_1 - 1}) & 0& 0 & \dots & 0 \\

\Phi^2_{\theta}(t_{N_1}) & 0 &\Phi_{c_2}(t_{N_1})  & 0&\dots& 0\\
\Phi^2_{\theta}(t_{N_1+1}) & 0 &\Phi_{c_2}(t_{N_1+1})  & 0& \dots&0\\
\vdots  &0& \vdots  & 0& \dots&0\\
\Phi^2_{\theta}(t_{N_2-1}) & 0 &\Phi_{c_2}(t_{N_2-1})  & 0& \dots & 0\\
\Phi^3_{\theta}(t_{N_2}) & 0& 0 &\Phi_{c_3}(t_{N_2})  & \dots& 0\\
\Phi^3_{\theta}(t_{N_2+1}) & 0& 0 &\Phi_{c_3}(t_{N_2+1})  & \dots& 0\\
\vdots  &0&0& \vdots  & \dots& 0\\
\Phi^3_{\theta}(t_{N_3-1}) & 0& 0 &\Phi_{c_3}(t_{N_3-1})  & \dots& 0\\
\vdots  &\vdots & \vdots & \vdots  & \ddots& 0\\
\Phi^T_{\theta}(t_{N_{T}}) & 0& 0 & 0& \dots& \Phi_{c_T}(t_{N_{T}}) \\\
\Phi^T_{\theta}(t_{N_{T}+1}) & 0& 0 & 0& \dots& \Phi_{c_T}(t_{N_{T}+1}) \\
\vdots  &\vdots & \vdots & \vdots  & \dots& \vdots\\
\Phi^T_{\theta}(t_{N}) & 0& 0 & 0& \dots& \Phi_{c_T}(t_{N}) \\
\end{bmatrix}\in \mathbb{R}^{(n \times N) \times (K+ m \times T)}
$$ 

Рассмотрим выражение непрерывности траектории на стыке соседних участков
$$
 \text{s.t.}\;\;\;\;\;\;\;\;  \forall j\in \{1, \dots, T-1\} \;\;\;\;\;\;\;\; 
x(t_{N_{j+1}}, \theta, c_j) - c_{j+1} =0   
$$
Варьируя парамеры $\theta$ и $c_j$ получаем $x(t_{N_{j+1}}, \theta + \delta \theta, c_j +  \delta c_{j}) - c_{j+1} -  \delta c_{j+1} =0 $
$$
x(t_{N_{j+1}}, \theta + \delta \theta, c_j + \delta c_j) - c_{j+1} - \delta c_{j+1} \approx x(t_{N_{j+1}}, \theta, c_j) - c_{j+1} + J^j_{\theta}(t_{N_{j+1}})\delta \theta + J_{c_j}(t_{N_{j+1}})\delta c_j - \delta c_{j+1} = 0
$$
 Данные выражения можно переписать в матричном виде
$$
G(p^k)\delta p^k - \mathbf{h^k} = 0
$$
где 
$$
G(p) =
\begin{bmatrix}
J^1_{\theta}(t_{N_2}) & J_{c_1}(t_{N_2}) & -I & 0& \dots& 0 & 0\\
J^2_{\theta}(t_{N_2})  & 0 & J_{c_2}(t_{N_3}) & -I & \dots& 0 & 0\\
\vdots  & \vdots  &  \vdots &  \vdots & \ddots &-I & 0\\
J^{T-1}_{\theta}(t_{N_T})  & 0 & 0 & 0 & \dots& J_{c_{T-1}}(t_{N_{T}})& -I\\
\end{bmatrix}\in \mathbb{R}^{(m \times (T-1)) \times (K+ m \times T)}
$$ 

$$ 
\mathbf{h} = 
\begin{bmatrix}
x(t_{N_{2}}, \theta, c_1) - c_{2} \\
x(t_{N_{3}}, \theta, c_2) - c_{3} \\
\vdots \\ 
x(t_{N_{T}}, \theta, c_{T-1}) - c_{T} \\
\end{bmatrix}\in \mathbb{R}^{m \times (T-1)} \quad
$$ 



Таким образом, задача Multiple Shooting эквивалентна задаче квадратичного программирования с ограничениями в виде равенств

$$
\min_{\delta p^k \in \mathbb{R}^{K + m \times T}} S(\delta p^k) = \| H(p^k)\delta p^k - \mathbf{r^k}\|_{\sigma}
$$

$$
 \text{s.t.}\;\;\;\;\;\;\;\;   G(p^k)\delta p^k - \mathbf{h^k} = 0
 $$

 $$
p^{k+1} = p^{k} + \delta p^k
$$

 

# Решение задачи Multiple shooting
$$
\min_{\delta p^k } S(\delta p^k) = \| H(p^k)\delta p^k - \mathbf{r^k}\|_{\sigma}
$$

$$
 \text{s.t.}\;\;\;\;\;\;\;\;   G(p^k)\delta p^k - \mathbf{h^k} = 0
$$
Зададим функцию Лагранжа

$$
L(\delta p^k, \lambda^k) = \| H(p^k)\delta p^k - \mathbf{r^k}\|_{\sigma} + (\lambda^k)^T(G(p^k)\delta p^k - \mathbf{h^k})
$$
где $\lambda^k \in R^{m \times (T-1)}$ — вектор множителей Лагранжа.
$$
\frac{dL(\delta p^k, \lambda^k)}{d\delta p^k} = 2H^TH\delta p^k - 2H^T\mathbf{r^k} + G^T\lambda^k = 0
$$
$$
\frac{dL(\delta p^k, \lambda^k)}{d\lambda^k} = G(p^k)\delta p^k - \mathbf{h^k} = 0
$$
Что можно переписать в виде системы линейных уравнений
$$
\begin{bmatrix}
2H^TH & G^T \\
G & 0 \\
\end{bmatrix}
\begin{bmatrix}
\delta p^k \\
\lambda^k \\
\end{bmatrix} =
\begin{bmatrix}
2H^T\mathbf{r^k} \\
\mathbf{h^k} \\
\end{bmatrix}
$$
