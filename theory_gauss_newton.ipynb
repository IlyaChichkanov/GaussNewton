# Постановка задачи оптимизации Single shooting

Пусть динамика системы задаётся дифференциальным уравнением

$$
\begin{equation}
    \dot{x} = f(t, x, \theta),\quad x(t_0) = c_0,
\end{equation}
$$

где 
- $t \in [t_0, t_f]$ — время,
- $x \in \mathbb{R}^{n_x}$ — вектор состояния системы,
- $\theta \in \mathbb{R}^{n_\theta}$ — вектор постоянных параметров,
- $c_0 \in \mathbb{R}^{n_x}$ — начальное состояние,
- $f: \mathbb{R}^{n_x} \times \mathbb{R}^{n_\theta} \to \mathbb{R}^{n_x}$ — непрерывно дифференцируема по $x$ и $\theta$.

Положим *неизвестными* и *независимыми* вектор параметров $\theta$ и начальное состояние $c_0$.

На временном интервале $[t_0, t_f]$ имеется $N$ измерений:

$$
y_i = h(x(t_i), \theta) + \varepsilon_i, \quad i = 1,\dots,N,
$$

где 
- $h: \mathbb{R}^{n_x} \times \mathbb{R}^{n_\theta} \to \mathbb{R}^{n_y}$ — функция измерений (непрерывно дифференцируема),
- $y_i \in \mathbb{R}^{n_y}$ — измерение,
- $\varepsilon_i \sim \mathcal{N}(0, \Sigma_i)$ — независимые гауссовские ошибки с ковариацией $\Sigma_i$.

**Цель:** найти $\hat\theta$ и $\hat c_0$, минимизирующие взвешенную сумму квадратов невязок.

Обозначим через $x(t \mid \theta, c_0)$ решение (1) в момент $t$ при заданных $\theta, c_0$.

Целевая функция:
$$
S(\theta, c_0) = \frac12 \sum_{i=1}^{N} \bigl( h(x(t_i \mid \theta, c_0), \theta) - y_i \bigr)^T W_i \bigl( h(x(t_i \mid \theta, c_0), \theta) - y_i \bigr),
$$
где $W_i = \Sigma_i^{-1}$ — весовые матрицы (обратные ковариациям шума).

Задача:
$$
\min_{\theta, c_0} S(\theta, c_0) \quad \text{при условии} \quad \dot{x} = f(t,x,\theta),\; x(t_0)=c_0.
$$


## Вариация целевой функции

Рассмотрим вариацию $S$ относительно $\theta$ и $c_0$.

Линеаризация решения:
$$
x(t \mid \theta+\delta\theta, c_0+\delta c_0) \approx x(t \mid \theta, c_0) + J_\theta(t) \delta\theta + J_{c_0}(t) \delta c_0,
$$
где
$$
J_\theta(t) = \frac{\partial x(t \mid \theta, c_0)}{\partial \theta} \in \mathbb{R}^{n_x \times n_\theta},\qquad
J_{c_0}(t) = \frac{\partial x(t \mid \theta, c_0)}{\partial c_0} \in \mathbb{R}^{n_x \times n_x}.
$$


## Вывод вариационных уравнений

Дифференцируем уравнение движения по $\theta$:

$$
\frac{\partial}{\partial \theta} \dot{x}(t) = \frac{\partial}{\partial \theta} f(t, x(t), \theta).
$$

Левая часть: меняем порядок дифференцирования

$$
\frac{\partial}{\partial \theta} \dot{x}(t) = \frac{d}{dt} \left( \frac{\partial x(t)}{\partial \theta} \right) = \dot{J}_\theta(t).
$$

Правая часть: по правилу дифференцирования сложной функции

$$
\frac{\partial}{\partial \theta} f(t, x(t), \theta) = \frac{\partial f}{\partial x} \cdot \frac{\partial x(t)}{\partial \theta} + \frac{\partial f}{\partial \theta}
= f_x(t,x,\theta)\, J_\theta(t) + f_\theta(t,x,\theta).
$$

Получаем дифференциальное уравнение:

$$
\dot{J}_\theta(t) = f_x(t,x,\theta)\, J_\theta(t) + f_\theta(t,x,\theta).
$$

Начальное условие: $x(t_0)=c_0$ не зависит от $\theta$, поэтому

$$
J_\theta(t_0) = \frac{\partial c_0}{\partial \theta} = \mathbf{0}_{n \times p}.
$$

### 2. Чувствительность по начальным условиям $c_0$

Определим

$$
J_{c_0}(t) = \frac{\partial x(t)}{\partial c_0} \in \mathbb{R}^{n \times n}.
$$

Дифференцируем уравнение по $c_0$:

$$
\frac{\partial}{\partial c_0} \dot{x}(t) = \frac{\partial}{\partial c_0} f(t, x(t), \theta).
$$

Левая часть:

$$
\frac{\partial}{\partial c_0} \dot{x}(t) = \frac{d}{dt} \left( \frac{\partial x(t)}{\partial c_0} \right) = \dot{J}_{c_0}(t).
$$

Правая часть: $f$ зависит от $c_0$ только через $x(t)$

$$
\frac{\partial}{\partial c_0} f(t, x(t), \theta) = \frac{\partial f}{\partial x} \cdot \frac{\partial x(t)}{\partial c_0} = f_x(t,x,\theta)\, J_{c_0}(t).
$$

Отсюда

$$
\dot{J}_{c_0}(t) = f_x(t,x,\theta)\, J_{c_0}(t).
$$

Начальное условие: $x(t_0)=c_0$, следовательно

$$
J_{c_0}(t_0) = \frac{\partial c_0}{\partial c_0} = I_n.
$$

### 3. Итоговые вариационные уравнения

Собирая полученные результаты:

$$
\begin{aligned}
\frac{d}{dt} J_\theta &= f_x(t,x,\theta)\, J_\theta + f_\theta(t,x,\theta), \quad & J_\theta(t_0) = 0,\\[4pt]
\frac{d}{dt} J_{c_0} &= f_x(t,x,\theta)\, J_{c_0}, \quad & J_{c_0}(t_0) = I_{n_x}.
\end{aligned}
$$

Эти уравнения позволяют численно вычислять производные решения по параметрам и начальным условиям, что необходимо при решении задач оптимизации (идентификация параметров, MHE) методом Гаусса–Ньютона.


Линеаризация функции измерений:
$$
h\bigl(x(t)+\delta x, \theta+\delta\theta\bigr) \approx h(x(t),\theta) + h_x \delta x + h_\theta \delta\theta,
$$
где $h_x = \frac{\partial h}{\partial x}$, $h_\theta = \frac{\partial h}{\partial \theta}$.

Подставляя $\delta x = J_\theta\delta\theta + J_{c_0}\delta c_0$, получаем:
$$
h(\cdot) \approx h(x(t),\theta) + \bigl( h_x J_\theta + h_\theta \bigr)\delta\theta + h_x J_{c_0}\delta c_0.
$$

Тогда невязка:
$$
r_i(\theta,c_0) = y_i - h(x(t_i),\theta).
$$

Вариация целевой функции (с весами):
$$
S(\theta+\delta\theta,c_0+\delta c_0) \approx \frac12 \sum_{i=1}^N \bigl\| \bigl( h_x J_\theta + h_\theta \bigr)\delta\theta + h_x J_{c_0}\delta c_0 - r_i \bigr\|_{W_i}^2.
$$

## Метод Гаусса-Ньютона

На $k$-й итерации имеем $\theta^k$, $c_0^k$. Решаем линейную задачу наименьших квадратов:

$$
\min_{\delta\theta,\delta c_0} \; \frac12 \sum_{i=1}^N \bigl\| A_i \delta\theta + B_i \delta c_0 - r_i^k \bigr\|_{W_i}^2,
$$

где
$$
A_i = h_x J_\theta(t_i) + h_\theta,\qquad B_i = h_x J_{c_0}(t_i).
$$

В матричной форме, вводя $p = [\theta;\, c_0] \in \mathbb{R}^{n_\theta + n_x}$ и $\delta p = [\delta\theta;\,\delta c_0]$:
$$
\min_{\delta p} \; \frac12 \| H(p^k) \delta p - r(p^k) \|_W^2,
$$
где
$$
H(p) = \begin{bmatrix} A_1 & B_1 \\ \vdots & \vdots \\ A_N & B_N \end{bmatrix} \in \mathbb{R}^{(n_y N) \times (n_\theta + n_x)},\qquad
r(p) = \begin{bmatrix} r_1 \\ \vdots \\ r_N \end{bmatrix} \in \mathbb{R}^{n_y N},
$$
а $W = \operatorname{blkdiag}(W_1,\dots,W_N)$ — блочно-диагональная весовая матрица.

Решение $\delta p^*$ даёт обновление:
$$
p^{k+1} = p^k + \delta p^*.
$$
Это стандартный метод Гаусса–Ньютона.

## Multiple shooting: расширение на несколько интервалов

Разобьём временной интервал на $T$ непересекающихся промежутков (шутов):
$$
[t_0,t_f] = \bigcup_{j=1}^{T} [\tau_{j-1}, \tau_j], \quad \tau_0=t_0,\; \tau_T=t_f.
$$
В каждом шуте $j$ своё начальное состояние $c_j \in \mathbb{R}^{n_x}$. Условия непрерывности:
$$
x(\tau_j \mid \theta, c_j) = c_{j+1}, \quad j=1,\dots,T-1.
$$

Обобщённый вектор параметров:
$$
p = [\theta;\, c_1; c_2; \dots; c_T] \in \mathbb{R}^{n_\theta + n_x T}.
$$

Невязки измерений: для каждого измерения $y_i$, принадлежащего шуту $j$,
$$
r_i = y_i - h(x(t_i \mid \theta, c_j), \theta).
$$

Линеаризация даёт блочную матрицу $H$:

$$
H(p) =
\begin{bmatrix}
\Phi^1_{\theta}(t_1) & \Phi_{c_1}(t_1) & 0 & \cdots & 0 \\
\vdots & \vdots & 0 & \cdots & 0 \\
\Phi^1_{\theta}(t_{N_1}) & \Phi_{c_1}(t_{N_1}) & 0 & \cdots & 0 \\
\Phi^2_{\theta}(t_{N_1+1}) & 0 & \Phi_{c_2}(t_{N_1+1}) & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
\Phi^T_{\theta}(t_{N}) & 0 & 0 & \cdots & \Phi_{c_T}(t_{N})
\end{bmatrix}
$$
где $\Phi^j_{\theta}(t) = h_x J_\theta^j(t) + h_\theta$, $\Phi_{c_j}(t) = h_x J_{c_j}(t)$.

Размерность: $H(p) \in \mathbb{R}^{(n_y N) \times (n_\theta + n_x T)}$.

Условия непрерывности в линеаризованном виде:
$$
G(p) \, \delta p = h_{\text{cont}}(p),
$$
где
$$
G(p) =
\begin{bmatrix}
J_\theta^1(\tau_1) & J_{c_1}(\tau_1) & -I & 0 & \cdots & 0 \\
J_\theta^2(\tau_2) & 0 & J_{c_2}(\tau_2) & -I & \cdots & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots \\
J_\theta^{T-1}(\tau_{T-1}) & 0 & 0 & 0 & \cdots & -I
\end{bmatrix},
$$
$$
h_{\text{cont}}(p) =
\begin{bmatrix}
c_2 - x(\tau_1 \mid \theta,c_1) \\
c_3 - x(\tau_2 \mid \theta,c_2) \\
\vdots \\
c_T - x(\tau_{T-1} \mid \theta,c_{T-1})
\end{bmatrix}.
$$
Размерность: $G(p) \in \mathbb{R}^{(n_x (T-1)) \times (n_\theta + n_x T)}$.

Таким образом, задача Multiple Shooting сводится к квадратичному программированию с линейными ограничениями-равенствами:
$$
\min_{\delta p} \; \frac12 \| H(p^k) \delta p - r(p^k) \|_W^2 \quad \text{s.t.} \quad G(p^k) \delta p = h_{\text{cont}}(p^k).
$$

## Решение задачи Multiple shooting методом множителей Лагранжа

Функция Лагранжа:
$$
L(\delta p, \lambda) = \frac12 \| H \delta p - r \|_W^2 + \lambda^T (G \delta p - h_{\text{cont}}).
$$

Условия оптимальности (ККТ):
$$
\begin{aligned}
\frac{\partial L}{\partial \delta p} &= H^T W (H \delta p - r) + G^T \lambda = 0,\\[4pt]
\frac{\partial L}{\partial \lambda} &= G \delta p - h_{\text{cont}} = 0.
\end{aligned}
$$

В матричной форме:
$$
\begin{bmatrix}
H^T W H & G^T \\
G & 0
\end{bmatrix}
\begin{bmatrix}
\delta p \\
\lambda
\end{bmatrix}
=
\begin{bmatrix}
H^T W r \\
h_{\text{cont}}
\end{bmatrix}.
$$

Эта система может быть решена, например, методом исключения или с помощью регуляризации (как в вашем коде). После нахождения $\delta p$ обновляем $p^{k+1} = p^k + \delta p$.

Таким образом, multiple shooting позволяет эффективно оценивать параметры и начальные условия на каждом интервале, обеспечивая непрерывность траектории через ограничения.

## Связь с кодом

- **`SystemJacobian`** вычисляет $f$, $h$, $J_\theta$, $J_{c_0}$.
- **`get_jacobian_solution`** интегрирует расширенную систему и возвращает $x(t)$, $J_\theta(t)$, $J_{c_0}(t)$.
- **`MultipleShooting.solve_batch`** собирает матрицы $H$ и $G$ и невязки $r$, $h_{\text{cont}}$ для одного батча.
- **Основной цикл** решает систему ККТ (с регуляризацией) и обновляет $\theta$ и $c_j$.

Это точная реализация описанного выше метода Гаусса–Ньютона с multiple shooting.

## Регуляризация

При решении задачи идентификации параметров методом Гаусса–Ньютона часто возникает проблема **плохой обусловленности** матрицы $H^T W H$ (особенно когда параметров много или данные содержат шум). Это приводит к неустойчивости оценок и переобучению.

### Байесовская интерпретация

Введём априорное распределение на параметры:
$$
\theta \sim \mathcal{N}(\bar\theta, \Lambda^{-1}), \qquad c_j \sim \mathcal{N}(\bar c_j, \Gamma_j^{-1}),
$$
где $\Lambda$, $\Gamma_j$ — матрицы точности (обратные ковариации). Тогда логарифм апостериорной плотности добавляет регуляризующие слагаемые:
$$
\frac12 \|\theta - \bar\theta\|_{\Lambda}^2 + \frac12 \sum_{j=1}^{T} \|c_j - \bar c_j\|_{\Gamma_j}^2.
$$

### Регуляризация в методе Гаусса–Ньютона

Целевая функция с регуляризацией:
$$
\min_{\delta p} \; \frac12 \| H \delta p - r \|_W^2 + \frac12 \| \delta p \|_{\text{diag}(\lambda)}^{2},
$$
где $\lambda$ — вектор коэффициентов регуляризации (обычно $\lambda_i \ge 0$). В матричной форме:
$$
\min_{\delta p} \; \frac12 \bigl( (H \delta p - r)^T W (H \delta p - r) + \delta p^T \operatorname{diag}(\lambda) \delta p \bigr).
$$

Это эквивалентно добавлению к матрице $H^T W H$ диагональной матрицы $\operatorname{diag}(\lambda)$:
$$
(H^T W H + \operatorname{diag}(\lambda)) \delta p = H^T W r.
$$


### Регуляризация в задаче с ограничениями (Multiple Shooting)

Для задачи с ограничениями-равенствами система ККТ преобразуется к виду:
$$
\begin{bmatrix}
H^T W H + \operatorname{diag}(\lambda) & G^T \\
G & 0
\end{bmatrix}
\begin{bmatrix}
\delta p \\ \lambda_{\text{lag}}
\end{bmatrix}
=
\begin{bmatrix}
H^T W r \\ h_{\text{cont}}
\end{bmatrix}.
$$


## Метод расширенного Лагранжиана (регуляризация блока ограничений)

При решении задачи multiple shooting методом Гаусса–Ньютона мы приходим к системе ККТ:

$$
\begin{bmatrix}
H^T W H & G^T \\
G & 0
\end{bmatrix}
\begin{bmatrix}
\delta p \\ \lambda
\end{bmatrix}
=
\begin{bmatrix}
H^T W r \\ h_{\text{cont}}
\end{bmatrix}.
$$

Блок \(0\) в правом нижнем углу делает матрицу плохо обусловленной, что затрудняет численное решение. Для улучшения сходимости часто применяют **метод расширенного Лагранжиана** (Augmented Lagrangian, метод Hestenes–Powell).

### Идея метода

Вместо исходной задачи решается последовательность вспомогательных задач, в которых к целевой функции добавляется квадратичный штраф за невыполнение ограничений:

$$
\min_{\delta p} \; \frac12 \| H \delta p - r \|_W^2 + \frac{\mu}{2} \| G \delta p - h_{\text{cont}} \|^2 \quad \text{без ограничений}.
$$

Условия оптимальности для этой задачи:

$$
H^T W (H \delta p - r) + \mu G^T (G \delta p - h_{\text{cont}}) = 0.
$$

Вводя множитель Лагранжа $\lambda = \mu (G \delta p - h_{\text{cont}}))$, можно переписать систему в виде:

$$
\begin{bmatrix}
H^T W H + \mu G^T G & G^T \\
G & -\frac{1}{\mu} I
\end{bmatrix}
\begin{bmatrix}
\delta p \\ \lambda
\end{bmatrix}
=
\begin{bmatrix}
H^T W r \\ h_{\text{cont}}
\end{bmatrix}.
$$

При $(\mu \to 0)$ эта система аппроксимирует исходную. На практике выбирают фиксированное $(\mu > 0)$ и решают систему, а затем итеративно уточняют множители.

### Упрощённая форма (как в вашем коде)

В вашей реализации используется **упрощённая регуляризация**:

$$
\begin{bmatrix}
H^T W H & G^T \\
G & -\mu I
\end{bmatrix}
\begin{bmatrix}
\delta p \\ \lambda
\end{bmatrix}
=
\begin{bmatrix}
H^T W r \\ h_{\text{cont}}
\end{bmatrix}.
$$

Здесь $(-\mu I)$ заменяет $(-\frac{1}{\mu}I)$ (при перенормировке строк). Такая замена делает матрицу лучше обусловленной и позволяет решать систему прямыми методами.

### Обновление множителей Лагранжа

Из второго блока уравнений получаем:

$$
G \delta p - h_{\text{cont}} = \mu \lambda.
$$

Но в классическом методе множители обновляются по формуле:

$$
\lambda_{\text{new}} = \lambda_{\text{old}} + \frac{1}{\mu} (G \delta p - h_{\text{cont}}).
$$
